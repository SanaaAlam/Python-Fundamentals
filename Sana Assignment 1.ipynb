```python
#(i) Create a Tuple of 10 sports names, do all operations addition,remove, change elements in the Tuple.
sports=("football","basketball","cricket","boxing","badminton","tennis","tabletennis","chess","archery","karate")
print(sports)
```

```python
#convert to list to perform calculations
sports_list=list(sports)
print(sports_list)
#adding elements
sports_list.append("volleyball")
print(sports_list)
to_extend=["hockey","volleyball1"]
sports_list.extend(to_extend)
print(sports_list)
sports_list.insert(5,"golf")
print(sports_list)
```

```python
#removing elements
sports_list.remove("volleyball1")
print(sports_list)
sports_list.pop(4)
print(sports_list)
```

```python
#change elements
sports_list[2]="swimming"
print(sports_list)
```

```python
#convert back to tuple
updated_sports=tuple(sports_list)
print(updated_sports)
```

```python
# (ii) Interchange List Items With & Without Temporary Variable
#using temporary values
l=[7,8]
print("list before interchanging:",l)
t=l[0]
l[0]=l[1]
l[1]=t
print("list after interchanging:",l)
```

```python
#without using temporary values
l1=[67,9]
print("list before interchanging:",l1)
l1[0],l1[1]=l1[1],l1[0]
print("list after interchanging:",l)
```

```python
# Create a dictionary
student = {
    "name": "Alex",
    "age": 21,
    "course": "Data Science"
}

# Add elements
student["grade"] = "A"
student["city"] = "New York"

# Remove an element
del student["age"]

# Copy dictionary
student_copy = student.copy()

# Create nested dictionary
student["scores"] = {
    "math": 95,
    "science": 88,
    "english": 92
}

print("Original Dictionary:", student)
print("Copied Dictionary:", student_copy)
```
