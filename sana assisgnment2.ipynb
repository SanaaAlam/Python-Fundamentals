```python

#prime check
#(iii) Create a Python module that contains a function to calculate the factorial of a number using recursion.  
n=int(input("enter a number: "))
if n<=1:
    print("not a prime number")
for i in range(2,int(n**0.5)+1):
    if n%i==0:
        print("not a prime number")
        break
else:
    print("prime number")
```

```python
#palindrome check
text=input("enter a string or a number:")
reversed_text=""
i=len(text)-1
while i>=0:
    reversed_text+=text[i]
    i-=1
if text==reversed_text:
    print("palindrome")
else:
    print("not a palindrome")
```

```python
#armstrong number check
num=int(input('enter a number:'))
temp=num
digits=[]
while temp>0:
    digits.append(temp%10)
    temp//=10
power=len(digits)
sum_of_powers=0
i=0
while i<power:
    sum_of_powers+=digits[i]**power
    i+=1

if sum_of_powers==num:
    print('armstrong number')
else:
    print('not an armstrong number')
```

```python
#fibonacci series
terms=int(input('enter number of terms'))
a,b=0,1
count=0
print("fibonacci series:")
while count<=terms:
    print(a,end=" ")
    a,b=b,a+b
    count+=1
          
```
